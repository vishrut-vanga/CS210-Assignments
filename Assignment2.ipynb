## Question 5 [24 points]
Write a function `getPercentageLate(df, airports = None)` that takes a DataFrame as input (the DataFrame is assumed to be formatted in the same way as `/shared/Datasets/Nov23_East_West_Flights.csv`) and a list of airports codes. If the airports are not specified, all airports are considered.

(Hint: you may have to create new columns and helper functions)

The function should return a DataFrame, grouped by the the following columns
* Airport
* Time of planned Departure as `TimeOfDay`, where the time of Departure (CRSDepTime) is grouped as (inclusive,exclusive):
    * Early Morning: 2am to 6am
    * Morning: 6am to 10am
    * Midday: 10am to 2pm
    * Afternoon: 2pm to 6pm
    * Evening: 6pm to 10pm
    * Late Night: 10pm to 2am  

with the following value
* NumberLateDeparture: the number of flights that left late in the group
* TotalDeparture: the total number of flights that left in the group
* PercentageLateDeparture: the percentage of flights that left late in the group (rounded to 2 decimal)

Your result DataFrame should be ordered by TimeOfDay, not alphabetical, but from Early Morning to Late Night. (hint, use pd.Categorical as a type for TimeOfDay)


````python
from A2sub import getPercentageLate

import csv
import numpy as np
import pandas as pd

from typing import List, Dict, Tuple

def getPercentageLate(df: pd.DataFrame, airports: List[str]= None) -> pd.DataFrame:
    df.loc[((df['CRSDepTime'] >= 200.0) & (df['CRSDepTime'] < 600.0)), 'TimeOfDay'] = 'Early Morning'
    df.loc[((df['CRSDepTime'] >= 600.0) & (df['CRSDepTime'] < 1000.0)), 'TimeOfDay'] = 'Morning'
    df.loc[((df['CRSDepTime'] >= 1000.0) & (df['CRSDepTime'] < 1400.0)), 'TimeOfDay'] = 'Midday'
    df.loc[((df['CRSDepTime'] >= 1400.0) & (df['CRSDepTime'] < 1800.0)), 'TimeOfDay'] = 'Afternoon'
    df.loc[((df['CRSDepTime'] >= 1800.0) & (df['CRSDepTime'] < 2200.0)), 'TimeOfDay'] = 'Evening'
    df.loc[((df['CRSDepTime'] >= 2200.0) | (df['CRSDepTime'] < 200.0)), 'TimeOfDay'] = 'Late Night'
    totallist = []
    for i in df.Origin.unique():
        em_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Early Morning')]
        emdf = em_df[em_df['DepDelay'] > 0.0]
        if len(emdf) != 0 and len(em_df) != 0:
            emlist = [1, i, len(emdf), len(em_df), round((len(emdf)/len(em_df))*100, 2)]
            totallist.append(emlist)
        m_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Morning')]
        mdf = m_df[m_df['DepDelay'] > 0.0]
        if len(mdf) != 0 and len(m_df) != 0:
            mlist = [2, i, len(mdf), len(m_df), round((len(mdf)/len(m_df))*100, 2)]
            totallist.append(mlist)
        mid_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Midday')]
        middf = mid_df[mid_df['DepDelay'] > 0.0]
        if len(middf) != 0 and len(mid_df) != 0:
            midlist = [3, i, len(middf), len(mid_df), round((len(middf)/len(mid_df))*100, 2)]
            totallist.append(midlist)
        a_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Afternoon')]
        adf = a_df[a_df['DepDelay'] > 0.0]
        if len(adf) != 0 and len(a_df) != 0:
            alist = [4, i, len(adf), len(a_df), round((len(adf)/len(a_df))*100, 2)]
            totallist.append(alist)
        e_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Evening')]
        edf = e_df[e_df['DepDelay'] > 0.0]
        if len(edf) != 0 and len(e_df) != 0:
            elist = [5, i, len(edf), len(e_df), round((len(edf)/len(e_df))*100, 2)]
            totallist.append(elist)
        ln_df = df[(df['Origin']==i) & (df['TimeOfDay']=='Late Night')]
        lndf = ln_df[ln_df['DepDelay'] > 0.0]
        if len(lndf) != 0 and len(ln_df) != 0:
            lnlist = [6, i, len(lndf), len(ln_df), round((len(lndf)/len(ln_df))*100, 2)]
            totallist.append(lnlist)
    totallist = sorted(totallist, key=lambda x: [x[0], x[1]])
    entire_Df = pd.DataFrame(totallist, columns=["TimeOfDay", "Origin", "NumberLateDeparture", "TotalDeparture", "PercentageLateDeparture"])
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(1, 'Early Morning')
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(2, 'Morning')
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(3, 'Midday')
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(4, 'Afternoon')
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(5, 'Evening')
    entire_Df['TimeOfDay'] = entire_Df['TimeOfDay'].replace(6, 'Late Night')
    #print(entire_Df.head())
    if airports == None:
        return entire_Df
    else:
        listofairlines = []
        for i in airports:
                newdf = entire_Df[entire_Df['Origin']==i]
                newdf.loc[ newdf['TimeOfDay'] == 'Early Morning', 'TimeOfDay'] = 1
                newdf.loc[ newdf['TimeOfDay'] == 'Morning', 'TimeOfDay'] = 2
                newdf.loc[ newdf['TimeOfDay'] == 'Midday', 'TimeOfDay'] = 3
                newdf.loc[ newdf['TimeOfDay'] == 'Afternoon', 'TimeOfDay'] = 4
                newdf.loc[ newdf['TimeOfDay'] == 'Evening', 'TimeOfDay'] = 5
                newdf.loc[ newdf['TimeOfDay'] == 'Late Night', 'TimeOfDay'] = 6
                listofairlines.append(newdf)
        c_df = pd.concat(listofairlines)
        c_df = c_df.sort_values(by='TimeOfDay')
        c_df.loc[ c_df['TimeOfDay'] == 1, 'TimeOfDay'] = 'Early Morning'
        c_df.loc[ c_df['TimeOfDay'] == 2, 'TimeOfDay'] = 'Morning'
        c_df.loc[ c_df['TimeOfDay'] == 3, 'TimeOfDay'] = 'Midday'
        c_df.loc[ c_df['TimeOfDay'] == 4, 'TimeOfDay'] = 'Afternoon'
        c_df.loc[ c_df['TimeOfDay'] == 5, 'TimeOfDay'] = 'Evening'
        c_df.loc[ c_df['TimeOfDay'] == 6, 'TimeOfDay'] = 'Late Night'
        c_df = c_df.reset_index(drop=True)
        return c_df
#getPercentageLate(nov23Flights)
getPercentageLate(nov23Flights, ['JFK'])
````
