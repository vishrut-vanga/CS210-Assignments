### Question 1 [10 points]
Write a program that reads words from a **filename**, which is given as
a string argument. (We will test your function with different files). It should **return** the words from the file in a list, sorted alphabetically in reverse according to the **last character** (and then the second to last, etc).

For instance, if the file has

```
bell tea  Zebra
apple
yellow
```

Then the output should be

`['tea', 'Zebra', 'apple', 'bell', 'yellow']`

````python
# YOUR CODE HERE
def reverse_sorted_words(filename: str) -> list[str]:
    text1 = open(filename)
    contents = text1.readlines()
    newlist = []
    for i in range(0, len(contents)):
        contents[i] = contents[i].strip('\n')
        contents[i] = contents[i].split(' ')
    for i in range(0, len(contents)):
        newlist.extend(contents[i])
    newlist = sorted(newlist, key= lambda x: x[::-1])
    return newlist

reverse_sorted_words("words.txt")
````
